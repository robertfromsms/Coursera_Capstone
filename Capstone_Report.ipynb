## Business Problem

We examine two macro variables in this study: competition frequency and population. By using clustering techniques, we can group Atlanta neighborhoods into competition/population combination categories. Lastly, we provide recommendations base on these combination categories. Of course, high population and low competition neighborhood is the most ideal location to establish the tacos and burritos restaurant. At the end, we aim to provide recommendations of best neighborhoods for our tacos and burritos restaurant.

## Methodology

In this study, I roughly examined the geographic area of each neighborhood, and decided that a venue search area of 1000 meters was appropriate. I performed API calls for all venues, then just food venues. After obtaining a list of venues corresponding to each Atlanta neighborhood, I excluded Atlanta neighborhoods with low venues, low food venues. Base on the FourSquare venue categories page, Mexican food venues would include all food venues with the categories: Mexican Restaurant, Botanero, Burrito Place, Taco Place, Tex-Mex Restaurant, Yucatecan Restaurant. I calculated the relevant competition, which is the percentage of Mexican food venues out of all food venues. Together with the 2010 population from the wikipedia page, I have all the relevant information.

Before performing k-means clustering, I standardized the data by removing the mean and scaling to unit variance. Then I performed elbow method analysis to determine the number of clusters. The results showed k should be 4 or 5. With the cluster label from the k-means clustering, I scatter-plotted each cluster in population vs. competition configuration space with different colors. It showed only 3 clusters are of interest. Clusters of neighborhoods that have low population and median/high competition are saturated. Finally, I examined the neighborhoods of the interesting clusters geographically for final analysis and insights.

## Analysis and Recommendations

Intuitively, we want higher population and lower competition. So in Fig 2, the flatter the slope of a line connecting the neighborhood to the origin, the less the risk. However, lower risk does not necessarily translate to amount of reward/income. The two neighborhoods (Midtown and Downtown) in Cluster 2 are the most populous neighborhoods, and they mark two extremely attractive areas in Atlanta. Downtown area has only two interesting neighborhoods (Downtown and Peachtree Heights West).

Midtown area, on the other hand, includes 3 other interesting neighborhoods. Home Park has low population and low competition, but it is right next to Midtown. Midtown has high population and median competition. Georgia Tech has median population and low competition. With 3 favorable neighborhoods next to each other, the Midtown area is better than the Downtown area. 

My final Recommendations for locations to establish a taco and burrito restaurant in Atlanta are Home Park and Midtown. Home Park would be less risky, because it is in a low competition, while low population, neighborhood, but it is right next to Midtown, a high population neighborhood. Midtown would be more risky, because being directly in the fire, so to speak, the upkeep cost must be high, however it is a high population neighborhood with favorable competition.
