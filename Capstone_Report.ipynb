# Introduction

Tacos, burritos and other Mexican dishes have broken through into the palate of mainstream America. Tacos and burritos are enjoyed by mainstream Americans, and of course the enormous amount of Mexican immigrants and ethnically and/or cultural Mexican Americans. While not an original idea, establishing a tacos and burritos restaurant is a tried and true concept. In this following report, we pay close attention to determining locations to establish said tacos and burritos restaurant in Atlanta, Georgia.

# Business Problem

We examine two macro variables in this study: competition frequency and population. By using clustering techniques on both problems, we can group Atlanta neighborhoods into competition categories and population categories. Lastly, we provide recommendations base on the combinations of competition and population. Of course, high population and low competition neighborhood is the most ideal location to establish the tacos and burritos restaurant. At the end, we aim to provide recommendations of best neighborhoods for our tacos and burritos restaurant.

# Target Audience

Entrepreneurs interested in establishing a taco and burrito restaurant in Atlanta, Georgia.
